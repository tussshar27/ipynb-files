Slicing                   
using .loc() and .iloc() functions.
1. .loc(): works on labeled index.
2. .iloc(): works on internal pandas numeric index.

```python
#there is a better and optimised way of slicing is by using .loc() and .iloc() functions.
#syntax of loc and iloc are both different.
```

```python
d1.loc[0:5,['Visibility (km)','Rel Hum (%)']]
#6 rows
```

```python
d1[['Visibility (km)','Rel Hum (%)']].loc[0:5]  #labeled index
#6 rows
```

```python
d1[['Visibility (km)','Rel Hum (%)']].iloc[0:5] #numeric index
#5 rows
```

```python
d1[['Visibility (km)','Rel Hum (%)']][0:6:2]
```

```python
d1[['Visibility (km)','Rel Hum (%)']].loc[0:5:2]
```

```python
d1[['Visibility (km)','Rel Hum (%)']].iloc[0:5:2]
```
